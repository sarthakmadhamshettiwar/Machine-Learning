```python
import numpy as np
import pandas as pd
```

```python
def cost_func(X, y, theta):
    m = len(y)
    y_pred = np.dot(X,theta)

    return (1/2*m)*(np.sum((y-y_pred)**2))
```

```python
def gradient_descent(X, y, theta, alpha, iter):
    m = len(y)
    cost = []
    for i in range(iter):
        y_pred  = np.dot(X,theta)
        cost.append([cost_func(X,y,theta)])
        error = (np.dot(X.T, y_pred-y))/m
        theta = theta - alpha*error
        

    
    return theta,cost
    
```

```python
import pandas as pd
data = pd.read_csv('datad.txt')
```

```python
m = data.input.values.size
# add another dimension to accomodate the intercept term and set it to all ones
X = np.append(np.ones((m, 1)), data.input.values.reshape(m, 1), axis=1)
y = data.output.values.reshape(m, 1)

theta = np.zeros((2,1))
```

```python
theta,cost = gradient_descent(X, y, theta, alpha=0.001, iter=10000)

print("h(x) = {} + {}x1".format(str(round(theta[0, 0], 2)),
                                str(round(theta[1, 0], 2))))

print('')
print(cost)
```
